# First, take a look at how the CSS prf model works.

```python
dm = np.load('/Volumes/BAHAMUT/STIM/design_matrix.npy')

# put time dimension last for popeye
dm = np.moveaxis(dm,0,2)

# binarize
dm[dm<0.5] = 0
dm[dm>0.5] = 1

# remove fixation point
dm[49,88,:] = 0
dm[49,89,:] = 0
dm[50,88,:] = 0
dm[50,89,:] = 0

#revert y axis
dm = dm[::-1,:,:] # this is how popeye wants y dim (0 point is top of dm)

TR = 1.5

stimulus = VisualStimulus(stim_arr = dm,viewing_distance = 225, screen_width =  69.84,scale_factor = 1,
                          tr_length = TR,
                          dtype = np.short)
```

```python
import popeye.css as css
```

```python
model=css.CompressiveSpatialSummationModel(stimulus, utils.double_gamma_hrf)
```

Put in some parameters for prediction.

```python
model.hrf_delay=-.25

# X coordinate
x = -5.24
# Y coordinate
y = 2.58

# SD
sigma = 4.24
hrf_delay=-0.25


beta = 0.55
baseline = -0.88
```

Now plot some predictions, with different exponents.

```python
data = model.generate_prediction(x, y, sigma,1,beta,baseline)
data2 = model.generate_prediction(x, y, sigma,0.5,beta,baseline)
data3 = model.generate_prediction(x, y, sigma,0.01,beta,baseline)
data4 = model.generate_prediction(x, y, sigma,0.001,beta,baseline)



plt.plot(data)
plt.plot(data2)
plt.plot(data3)
plt.plot(data4)
```

Here, we see the compressive effect. Essentially, the effect of decreasing the exponent is such that the response is less dependent on the overlap betweem the stimulus and the prf

The way this works is that we first generate a predicted response from the overlap between the bar and the stimulus.

```python
from popeye.spinach import generate_og_receptive_field, generate_rf_timeseries_nomask


rf = generate_og_receptive_field(x, y, sigma,model.stimulus.deg_x0, model.stimulus.deg_y0)
        
        # normalize by the integral
rf /= ((2 * np.pi * sigma**2) * 1/np.diff(model.stimulus.deg_x0[0,0:2])**2)
        
        # extract the stimulus time-series
response = generate_rf_timeseries_nomask(model.stimulus.stim_arr0, rf)
```

```python
plt.plot(response)
```

Thus **importantly**, a zero response is predicted when the bar is outside of the rf - and the other responses are dependent on the spatial overlap.

At this point, the compression is applied.

```python
plt.plot(response/np.max(response))

plt.plot(response **0.05/np.max(response **0.05),'r')
plt.plot(response **0.001/np.max(response **0.001),'g')
```

The smaller the exponent, the more we expect a large response to a stimulus at any location within the rf. Essentially, it will end up being a squarewave.

Next, this response is convolved with the hrf.

```python
from scipy.signal import fftconvolve

hrf = model.hrf_model(model.hrf_delay, model.stimulus.tr_length)
        
r1=fftconvolve(response**1, hrf)[0:len(response)]
r2=fftconvolve(response**0.05, hrf)[0:len(response)]
plt.plot(r1)
plt.plot(r2)
```

Then we just go through some additional stages to normalise the data

```python
r1norm=r1 = (r1 - np.mean(r1)) / np.mean(r1)
r2norm=r2 = (r2 - np.mean(r2)) / np.mean(r2)
plt.plot(r1norm)
plt.plot(r2norm)
```

And this is how we end up with what we had before.

This is how we make predictions with a css prf model. However, the cf model differs in a number of important ways.

1. The input to the compressive stage is the gaussian weighted signals in V1 (i.e. real data), not an arbitrary model response.
2. Therefore, we don't really predict a 'zero' response at any stage. The signal isn't going to be at 0, or even stationary when there is no stimulus.
3. We dont colvolve with a hrf.

Here we can see the compressive effect, but we lose values wherever the signal becomes negative.

This is because the square root of negatives are imaginary numbers. 

One thing that we could try is to just not transform the signal when it is negative, or falls below some threshold.

We know, for instance, that NOTHING is presented on the screen for the first and last 15 TRs. To my mind, it doesnt make sense to compress these signals - since they are 'non visual'.

This is at least how the css prf model treats periods of no stimulation.

Therefore, we define our threshold as the maximum response within the start and end periods (where there is no bar on the screen)

```python
thresh=np.max(np.concatenate([mydata[0:15],mydata[110:120]]))
```

Therefore, if these are forced to be below zero, then we can avoid transforming them altogether

```python
compdat=(mydata-thresh)**0.01

compdat[mydata<=thresh]=mydata[mydata<=thresh]
plt.plot(mydata)
plt.plot(compdat)
```

Thus, our compressive model just looks like this. 

Clearly this threshold isnt quite right, as we are not compressing aspects of the hrfs

So another solution would be to just ingore this first period (with no visual stimulation) altogether.

I'm not comfortable with ignoring the last period altogether, since there could be lingering visual responses left in there.

Since we are no longer including this first period, we can be more liberal about the threshold, and instead use the median within that period.

```python
thresh2=np.median(np.concatenate([mydata[0:15],mydata[110:120]]))
```

```python
plt.plot(mydata-thresh2)
```

```python
mydata2=np.copy(mydata-thresh2)

mydata3=np.copy(mydata2)
mydata3[15:110]=mydata3[15:110]**0.1

mydata3[15:110][mydata2[15:110]<0]=mydata2[15:110][mydata2[15:110]<0]
plt.plot(mydata2)
plt.plot(mydata3,'r')
```

This does seem a lot more reasonable, but of course, it will look increasingly crazy the lower the exponent is.

I guess it only looks crazy because of our linear way of thinking....

```python
def compress(signal,begin_vis,end_vis,exp,plot):
    thresh=np.median(np.concatenate([signal[:begin_vis],signal[end_vis:]]))
    base=signal-thresh
    signal2=np.copy(base)
    signal2[begin_vis:end_vis]=signal2[begin_vis:end_vis]**exp
    signal2[begin_vis:end_vis][base[begin_vis:end_vis]<0]=base[begin_vis:end_vis][base[begin_vis:end_vis]<0]
    if plot:
        plt.plot(base)
        plt.plot(signal2)
    
    return signal2
```

```python

exponents=[0.9,0.7,0.5,0.1,0.001]

for e in exponents:
    compress(predictions[9,0,:],15,110,e,True)
```

We could avoid thresholding at all - if we do this then small signal changes (like those at around 55 seconds) will be amplified.

Arguably this is 'the point' if they correspond to visual stimulation.

To do this however, we would need to force everything to be positive. But then, if we just apply a power transformation to positive data, then we end up with a function that is just a factor of the original function - so this would be pointless I think.
